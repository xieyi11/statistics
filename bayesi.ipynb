### A/B test 双样本对照z检验

https://blog.csdn.net/buracag_mc/article/details/74905483

```python
import numpy as np
n1=1000
a1=0.07
n2=1010
a2=0.08
```

```python
z=(a2-a1)/np.sqrt(a2*(1-a2)/n2+a1*(1-a1)/n1)
print('z=%2f '%z)
```

显著性水平α=0.05，z(1-α)=1.64

z<z(1-α)，指标a2并不一定比a1高
